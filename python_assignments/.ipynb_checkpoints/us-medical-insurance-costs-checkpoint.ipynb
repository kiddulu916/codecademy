# U.S. Medical Insurance Costs

```python
# Import necessary python libraries
import csv
import json
import pandas as pd
import math
```

```python
# Convert dataset csv file into json
# Store csv file path to a variable and store the filepath you want the 
# json file to be written to
csv_filepath = 'insurance.csv'
json_filepath = 'insurance.json'

# Function to convert csv to json
def csv_to_json(csv_filepath, json_filepath):
    data = [] # Empty list to hold dataset
    with open(csv_filepath, 'r') as csvfile:
        csv_reader = csv.DictReader(csvfile) # convert csv data to a dictionary
        for row in csv_reader:
            data.append(row)
    
    # Write data to the json file
    with open(json_filepath, 'w') as jsonfile:
        json.dump(data, jsonfile, indent=4)

# Call function to convert the dataset
csv_to_json(csv_filepath, json_filepath)
```

```python
# open json file and save dataset within a list to a variable 
with open(json_filepath, 'r') as file:
    dataset = json.load(file) 
```

```python
# Number of patients in the dataset
total_num_of_patients = len(dataset)
total_num_of_smokers = 0
total_num_of_nonsmokers = 0
total_num_with_children = 0
total_num_without_children = 0

# Variables to hold the ages of the youngest and oldest patient
youngest_patient = 100
oldest_patient = 0
num_of_male_patients = 0
num_of_female_patients = 0
num_of_male_smokers = 0
num_of_female_smokers = 0
num_of_males_with_children = 0
num_of_females_with_children = 0

for data in dataset:
    if data['smoker'] == 'yes':
        total_num_of_smokers += 1
    elif data['smoker'] == 'no':
        total_num_of_nonsmokers += 1

for data in dataset:
    if int(data['children']) == 0:
        total_num_with_children +=1
    elif int(data['children']) != 0:
        total_num_without_children +=1
    
# Iterate through the dataset updating the youngest_patient variable
# to find the youngest patient
for data in dataset:
    if int(data['age']) < youngest_patient:
        youngest_patient = int(data['age'])

# Iterate through the dataset updating the oldest_patient variable
# to find the oldest patient
for data in dataset:
    if int(data['age']) > oldest_patient:
        oldest_patient = int(data['age'])

for data in dataset:
    if data['sex'] == 'male':
        num_of_male_patients += 1
    elif data['sex'] == 'female':
        num_of_female_patients += 1

for data in dataset:
    if data['sex'] == 'male' and data['smoker'] == 'yes':
        num_of_male_smokers += 1
    elif data['sex'] == 'female' and data['smoker'] == 'yes':
        num_of_female_smokers += 1

for data in dataset:
    if data['sex'] == 'male' and int(data['children']) != 0:
        num_of_males_with_children += 1
    elif data['sex'] == 'female' and int(data['children']) != 0:
        num_of_females_with_children += 1

diff_in_total_smokers_nonsmokers = total_num_of_smokers - total_num_of_nonsmokers
diff_in_total_with_without_children = total_num_with_children - total_num_without_children
diff_in_males_to_females = num_of_male_patients - num_of_female_patients
diff_in male_to_female_smokers = num_of_male_smokers - num_of_female_smokers
diff_in_male_to_female_with_children = num_of_males_with_children - num_of_females_with_children

# Print out the results
print(f"Total number of patients: {total_num_of_patients}")
print(f"Patients ages range from: {youngest_patient} - {oldest_patient}\n")
print(f"There are {num_of_male_patients} male patients.\n{num_of_male_smokers} of those males are smokers and {num_of_males_with_children} of them have at least 1 child\n")
print(f"There are {num_of_female_patients} female patients.\n{num_of_female_smokers} of those females are smokers and {num_of_females_with_children} of them have at least 1 child\n")
```

```python
# Empty lists to hold all the ages from each of its corresponding columns
ages_list = []
smoker_age_list = []
nonsmoker_age_list = []
with_children_age_list = []
no_children_age_list = []

# Zeroed variables to hold the sum of all ages from each column
sum_of_ages = 0
sum_of_smoker_ages = 0
sum_of_nonsmoker_ages = 0
sum_of_with_children_ages = 0
sum_of_no_children_ages = 0

# Iterate through the dataset to add all ages to its list
# and sort the ages of smokers and nonsmokers to there own lists
for data in dataset:
    ages_list.append(data['age'])
    if data['smoker'] == 'yes':
        smoker_age_list.append(data['age'])
    elif data['smoker'] == 'no':
        nonsmoker_age_list.append(data['age'])
    else:
        break

# Iterate through the dataset to seperate the ages of patients with children
# and without children to there defined lists
for data in dataset:
    if data['children'] != '0':
        with_children_age_list.append(data['age'])
    elif data['children'] == '0':
        no_children_age_list.append(data['age'])
    else:
        break    

# Calculate the sum of each list
for age in ages_list:
    sum_of_ages += int(age)

for age in smoker_age_list:
    sum_of_smoker_ages += int(age)

for age in nonsmoker_age_list:
    sum_of_nonsmoker_ages += int(age)

for age in with_children_age_list:
    sum_of_with_children_ages += int(age)

for age in no_children_age_list:
    sum_of_no_children_ages += int(age)

# Calculate the averages of each list
age_average = sum_of_ages / len(ages_list)
smoker_age_average = sum_of_smoker_ages / len(smoker_age_list)
nonsmoker_age_average = sum_of_nonsmoker_ages / len(nonsmoker_age_list)
with_children_age_average = sum_of_with_children_ages / len(with_children_age_list)
no_children_age_average = sum_of_no_children_ages / len(no_children_age_list)

# Round the averages
rounded_age_average = int(age_average)
rounded_smoker_age_average = int(smoker_age_average)
rounded_nonsmoker_age_average = int(nonsmoker_age_average)
rounded_with_children_age_average = int(with_children_age_average)
rounded_no_children_age_average = int(no_children_age_average)

# Print out the results
print(f"Average age of patients overall: {rounded_age_average}\n")
print(f"Average age of patients that smoke: {rounded_smoker_age_average}\nAverage age of patients that don't smoke: {rounded_nonsmoker_age_average}\n")
print(f"Average age of patients with at least 1 child: {rounded_with_children_age_average}\nAverage age of patients with 0 children: {rounded_no_children_age_average}")
```

```python
# Find total average costs
# Empty list to store insurance costs 
insurance_costs = []
total_insurance_cost = 0

# Iterate through dataset and append charges to insurance_costs list
for data in dataset:
    insurance_costs.append(data['charges'])

# Iterate through insurance_costs list and add them all to total_insurance_costs
for cost in insurance_costs:
    total_insurance_cost += float(cost)

# Average cost of overall costs
average_total_cost = total_insurance_cost / len(insurance_costs)

# Shorten the decimal to only 2 decimal places    
rounded_total_cost = round(total_insurance_cost, 2)
rounded_average_cost = round(average_total_cost, 2)

print(f"The total amount spent on insurance is: ${rounded_total_cost} \nThe average insurance cost of overall costs is : ${rounded_average_cost}")
```

```python
# Find the differance in costs between males and females
# Empty lists to store all male and female charges
male_costs = []
female_costs = []

# Variables to add up all the charges
total_male_cost = 0
total_female_cost = 0

""" 
Iterate through dataset and seperate male costs 
and female costs to there own lists
"""
for data in dataset:
    if data['sex'] == 'male':
        male_costs.append(data['charges'])
    elif data['sex'] == 'female':
        female_costs.append(data['charges'])
    else:
        break

# Calculate totals for male and female costs
for cost in male_costs:
    total_male_cost += float(cost)

for cost in female_costs:
    total_female_cost += float(cost)

# Calculate averages of male and female costs
average_male_cost = total_male_cost / len(male_costs)
average_female_cost = total_female_cost / len(female_costs)

# Calculate differences between male and female total and average costs
difference_in_total_costs = total_male_cost - total_female_cost
difference_in_average_costs = average_male_cost - average_female_cost

# Round total, average, and differences to 2 decimal points
rounded_total_male_cost = round(total_male_cost, 2)
rounded_total_female_cost = round(total_female_cost, 2)

rounded_average_male_cost = round(average_male_cost, 2)
rounded_average_female_cost = round(average_female_cost, 2)

rounded_total_difference = round(difference_in_total_costs, 2)
rounded_average_difference = round(difference_in_average_costs, 2)

print(f"Total insurance costs for males is: ${rounded_total_male_cost}0\nTotal insurance costs for females is: ${rounded_total_female_cost}\n")
print(f"Difference of total costs between males and females is: ${rounded_total_difference}0\n")
print(f"Average insurance costs for males is: ${rounded_average_male_cost}\nAverage insurance costs for females is: ${rounded_average_female_cost}\n")
print(f"Difference of average costs between males and females is: ${rounded_average_difference}\n")
    
```

```python
# Find the differance in cost between smokers and non-smokers
# Empty lists to store all smoker and non-smoker charges
smoker_costs = []
nonsmoker_costs = []
male_smoker_costs = []
female_smoker_costs = []
male_nonsmoker_costs = []
female_nonsmoker_costs = []

# Variables to add up all the charges
total_smoker_cost = 0
total_nonsmoker_cost = 0
total_male_smoker_cost = 0
total_female_smoker_cost = 0
total_male_nonsmoker_cost = 0
total_female_nonsmoker_cost = 0

""" 
Iterate through dataset and seperate smoker costs 
and non-smoker costs to there own lists
"""
for data in dataset:
    if data['smoker'] == 'yes':
        smoker_costs.append(data['charges'])
    elif data['smoker'] == 'no':
        nonsmoker_costs.append(data['charges'])
    else:
        break

for data in dataset:    
    if data['smoker'] == 'yes' and data['sex'] == 'male':
        male_smoker_costs.append(data['charges'])
    elif data['smoker'] == 'yes' and data['sex'] == 'female':
        female_smoker_costs.append(data['charges'])
    else:
        break
        
for data in dataset:
    if data['smoker'] == 'no' and data['sex'] == 'male':
        male_nonsmoker_costs.append(data['charges'])
    elif data['smoker'] == 'no' and data['sex'] == 'female':
        female_nonsmoker_costs.append(data['charges'])
    else:
        break

# Calculate totals for smokers and non-smokers costs
for cost in smoker_costs:
    total_smoker_cost += float(cost)

for cost in nonsmoker_costs:
    total_nonsmoker_cost += float(cost)

for cost in male_smoker_costs:
    total_male_smoker_cost += float(cost)

for cost in male_nonsmoker_costs:
    total_male_nonsmoker_cost += float(cost)

for cost in female_smoker_costs:
    total_female_smoker_cost += float(cost)

for cost in female_nonsmoker_costs:
    total_female_nonsmoker_cost += float(cost)

# Calculate averages of smokers and nonsmoker costs
average_smoker_cost = total_smoker_cost / len(smoker_costs)
average_nonsmoker_cost = total_nonsmoker_cost / len(nonsmoker_costs)

average_male_smoker_cost = total_male_smoker_cost / len(male_smoker_costs)
average_male_nonsmoker_cost = total_male_nonsmoker_cost / len(male_nonsmoker_costs)

average_female_smoker_cost = total_female_smoker_cost / len(female_smoker_costs)
average_female_nonsmoker_cost = total_female_nonsmoker_cost / len(female_nonsmoker_costs)

# Calculate differences between smoker and nonsmoker total and average costs
difference_in_smoker_total_costs = total_smoker_cost - total_nonsmoker_cost
difference_in_smoker_average_costs = average_smoker_cost - average_nonsmoker_cost

difference_in_male_smoker_total_costs = total_male_smoker_cost - total_male_nonsmoker_cost
difference_in_male_smoker_average_costs = average_male_smoker_cost - average_male_nonsmoker_cost

difference_in_female_smoker_total_costs = total_female_smoker_cost - total_female_nonsmoker_cost
difference_in_female_smoker_average_costs = average_female_smoker_cost - average_female_nonsmoker_cost

difference_in_male_female_smoker_total_costs = total_male_smoker_cost - total_female_smoker_cost
difference_in_male_female_smoker_average_costs = average_male_smoker_cost - average_female_smoker_cost

difference_in_male_female_nonsmoker_total_costs = total_male_nonsmoker_cost - total_female_nonsmoker_cost
difference_in_male_female_nonsmoker_average_costs = average_male_nonsmoker_cost - average_female_nonsmoker_cost

# Round total, average, and differences to 2 decimal points
rounded_total_smoker_cost = round(total_smoker_cost, 2)
rounded_total_nonsmoker_cost = round(total_nonsmoker_cost, 2)
rounded_total_male_smoker_cost = round(total_male_smoker_cost, 2)
rounded_total_male_nonsmoker_cost = round(total_male_nonsmoker_cost, 2)
rounded_total_female_smoker_cost = round(total_female_smoker_cost, 2)
rounded_total_female_nonsmoker_cost = round(total_female_nonsmoker_cost, 2)

rounded_average_smoker_cost = round(average_smoker_cost, 2)
rounded_average_nonsmoker_cost = round(average_nonsmoker_cost, 2)
rounded_average_male_smoker_cost = round(average_male_smoker_cost, 2)
rounded_average_male_nonsmoker_cost = round(average_male_nonsmoker_cost, 2)
rounded_average_female_smoker_cost = round(average_female_smoker_cost, 2)
rounded_average_female_nonsmoker_cost = round(average_female_nonsmoker_cost, 2)

rounded_smoker_total_difference = round(difference_in_smoker_total_costs, 2)
rounded_smoker_average_difference = round(difference_in_smoker_average_costs, 2)
rounded_male_smoker_total_difference = round(difference_in_male_smoker_total_costs, 2)
rounded_male_smoker_average_difference = round(difference_in_male_smoker_average_costs, 2)
rounded_female_smoker_total_difference = round(difference_in_female_smoker_total_costs, 2)
rounded_female_smoker_average_difference = round(difference_in_female_smoker_average_costs, 2)
rounded_male_female_smoker_total_difference = round(difference_in_male_female_smoker_total_costs, 2)
rounded_male_female_smoker_average_difference = round(difference_in_male_female_smoker_average_costs, 2)
rounded_male_female_nonsmoker_total_difference = round(difference_in_male_female_nonsmoker_total_costs, 2)
rounded_male_female_nonsmoker_average_difference = round(difference_in_male_female_nonsmoker_average_costs, 2)


print(f"Total insurance costs for males is: ${rounded_total_male_cost}0\nTotal insurance costs for females is: ${rounded_total_female_cost}\n")
print(f"Difference of total costs between males and females is: ${rounded_total_difference}0\n")
print(f"Average insurance costs for males is: ${rounded_average_male_cost}\nAverage insurance costs for females is: ${rounded_average_female_cost}\n")
print(f"Difference of average costs between males and females is: ${rounded_average_difference}\n")
    
```
